```python
import json
import unidecode
import requests

# Open data
url = "https://raw.githubusercontent.com/AntoniaDahuabe/AntoniaDahuabe.github.io/refs/heads/main/Habitantes_Superficie_Chile.json"
response = requests.get(url)
data = response.json()

# Normalization of string
def normalize_string(s):
    return unidecode.unidecode(s.strip().lower())

# Iteration to normalize comuna name within each region dictionary
for region in data:
    if "comunas" in region:
        for comuna in region["comunas"]:
            comuna["NormalizedName"] = normalize_string(comuna["name"])

# Save new json
output_file = "processed_Habitantes_Superficie_Chile.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

# Download
from google.colab import files
files.download(output_file)

```

```python
# Open new json
with open("processed_Habitantes_Superficie_Chile.json", "r", encoding="utf-8") as f:
    processed_data = json.load(f)

# Print part of it to double check
for region in processed_data[:3]:
    print(json.dumps(region, indent=2, ensure_ascii=False))
```

```python
import json

# Load the JSON data
with open('processed_Habitantes_Superficie_Chile.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Extract communes into a flat list
communes = []
for region in data:
    for comuna in region['comunas']:
        communes.append(comuna)

# Now 'communes' is a list of dictionaries with keys including 'NormalizedName'
```

```python
# Save the flattened data to a JSON file
output_file = "flattened_Habitantes_Superficie_Chile.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(communes, f, ensure_ascii=False, indent=2)

# Download the file (only for Google Colab)
files.download(output_file)
```
